```python
"""
 Creating a figure and axis objects with a specified size
 
"""

def line_graph():
   
    fig, ax = plt.subplots(figsize=[12, 5])
    
    # Plotting gas prices for each country over the years
    
    # Plot UK gas prices
    ax.plot(df.Year, df.UK, color='r') 
    
    # Plot Germany gas prices
    ax.plot(df.Year, df.Germany, color='b')  
    
    # Plot France gas prices
    ax.plot(df.Year, df.France, color='g')  
    
    # Plot Italy gas prices
    ax.plot(df.Year, df.Italy, color='y')  
    
    # Set x-axis ticks to be every year
    plt.xticks(df.Year[::1])
    
    # Enable gridlines on the plot
    plt.grid(True)
    
    # Set the title of the plot
    plt.title('Top Four countries selling gas for high price')
    
    # Set label for the x-axis
    plt.xlabel('Years', fontdict={'fontsize': 12})
    
    # Set label for the y-axis
    plt.ylabel('Price', fontdict={'fontsize': 11})

    # Add a legend to the plot indicating which line corresponds to which country
    plt.legend(['UK', 'Germany', 'France', 'Italy'])
    plt.savefig('line.png')
  
    plt.show()
```

```python
"""
Function to generate a line graph.

"""

line_graph()
```

```python
"""
Creating and displaying a bar graph using matplotlib.

This function generates a bar graph to visualize gas prices for three countries (USA, Canada, Australia) over the years.

"""

def bar_graph():
    # Creating a figure and axis objects with a specified size
    fig, ax = plt.subplots(figsize=[13, 6])
    
    # Generating X-axis values for each year
    X_axis = np.arange(len(df.Year))
    
    # Enabling gridlines on the plot
    plt.grid(True)
    
    # Plotting the bars for each country's gas prices
    
    # Bars for USA
    ax.bar(X_axis + 0.2, df.USA, 0.2)
    
    # Bars for Canada
    ax.bar(X_axis + 0.4, df.Canada, 0.2)
    
    # Bars for Australia
    ax.bar(X_axis + 0.6, df.Australia, 0.2)
    
    # Setting X-axis ticks to represent each year
    plt.xticks(X_axis, df.Year)
    
    # Add a legend to the plot indicating which bar corresponds to which country
    plt.legend(['USA', 'Canada', 'Australia'])
    
    # Setting label for x-axis
    plt.xlabel('Years', fontdict={'fontsize': 15})
    
    # Setting label for y-axis
    plt.ylabel("Price", fontdict={'fontsize': 15})
    
    # Setting title of the plot
    plt.title('Top Three countries selling gas for Low price over the years', fontdict={'fontsize': 16})
    
    # Saving the plot as a PNG file
    plt.savefig('bar.png')
    
    # Displaying the plot
    plt.show()
```

```python
"""
Function to generate a bar graph.

"""
bar_graph()
```

```python
"""

Creating a function named box_plot

"""

def box_plot():
    
    # Create a new figure with a specified size
    plt.figure(figsize=[12, 7])
    
    # Set the font scale for seaborn plots
    sns.set(font_scale=1.2)
    
    # Creating a box plot for the gas prices of the multiple countries
    sns.boxplot(df[['Australia', 'Canada', 'France', 'Germany', 'Italy', 'Japan', 'Mexico', 'South Korea', 'UK', 'USA']])
    
    # Setting  label for  x-axis
    font = {'fontsize': 16}
    plt.xlabel('Countries', fontdict=font, color='r')
    
    # Setting label for  y-axis
    plt.ylabel('Price', fontdict=font, color='r')
    
    # Setting  title of the plot
    plt.title('Box Plot', fontdict=font, color='b')
    
    # Saving the plot as a PNG file
    plt.savefig('Box.png')
    
    # Displaying the plot
    plt.show()
```

```python
""" 
Function to generate a box plot.

"""
box_plot()
```
